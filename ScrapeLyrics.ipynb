# Import Packages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import datetime
```

# Produce List Of Dates For Webscrape

```python
Data = pd.read_csv("TopChartSongs.csv")
Data = Data[["Track","Artist"]].drop_duplicates(keep="first")
Data = Data.reset_index(drop = True)
Data.shape
```

```python
Lyrics = []
for i, j in enumerate(Data.Track):
    try:
        print(str(i), "- Pass")
        Website = "http://api.chartlyrics.com/apiv1.asmx/SearchLyricDirect?artist="+str(Data.Artist[i])+"&song="+str(Data.Track[i])
        r = requests.get(Website)
        Output = BeautifulSoup(r.text,"html.parser")
        Output = str(Output.find_all("lyric"))
        Lyrics.append(re.search("(?<=<lyric>).*(?=</lyric>)",Output,re.DOTALL)[0])
    except:
        Lyrics.append("Error")
        print(str(i) + "- Error")
    pass
```
